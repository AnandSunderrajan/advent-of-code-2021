Question 1:

```python
inputFile = open('input.txt', 'r')
depths = list(map(int, inputFile.read().splitlines()))
result = 0
for i in range(len(depths)-1):
    if depths[i] < depths[i+1]:
        result += 1 
print(result)
inputFile.close()
```

Question 2:

```python
inputFile = open('input.txt', 'r')
depths = list(map(int, inputFile.read().splitlines()))
result = 0
size = len(depths)
base = sum(depths[:3])
for i in range(3, size):
    next_sum = base + depths[i] - depths[i-3]
    if base < next_sum:
        result += 1 
    base = next_sum
print(result)
inputFile.close()
```
